```python
# full.
#
# ax,ay: anchor's location

def conv2d_same(src,kernel,ax,ay,stride):
    kw,kh=kernel.shape
    iwo,iho=src.shape
    if kw!=kh or iwo!=iho:
        print("kernel and image should be square")
        return 
    r=int(kw/2)
    dst=src.copy()
    dst_full=np.pad(dst,((r,r),(r,r)),constant_values=(0,0),mode='constant') # 2r0
    dst_full=dst_full.astype(np.float32)
    #dest = dst_full.copy()
    dest=np.zeros(dst_full.shape,np.float32)
    i=j=r
    iw,ih=dest.shape
    #print("r:%d,iw:%d,ih:%d,pad_iw:%d,pad_ih:%d" %(r,iwo,iho,iw,ih))
    
    # DEBUG: 
    #       True)  
    #       Flase) 
    
    DEBUG_EN=False#True
    if DEBUG_EN==True:
        debug_cnt=0
    else:
        debug_cnt=10
    
    if kw%2==1:
        odds=1
    else:
        odds=0
    #print("i:%d,iw:%d,ih:%d,r:%d,odds:%d" %(i,iw,ih,r,odds))
    sax=ax
    say=ay
    while i< (iw-2*r):#(iw+2*r-r-odds-1):
        while j<(ih-2*r):#(ih+2*r-r-odds-1):
            temp=kernel*dst_full[(i-r):(i+r+odds),(j-r):(j+r+odds)]
            
            if debug_cnt<10:
                print("[i:%d,j:%d]origin dest[%d,%d]:%d" %(i,j,ax,ay,dest[ax,ay]))
                print("kernel:",kernel)
                print("dst_full[%d:%d,%d:%d]:%s" %((i-r),(i+r+odds),(j-r),(j+r+odds),dst_full[(i-r):(i+r+odds),(j-r):(j+r+odds)]))
            
            dest[ax,ay]=np.sum(temp)
            
            if debug_cnt<10:
                debug_cnt+=1
                print("[i:%d,j:%d]after dest[%d,%d]:%d" %(i,j,ax,ay,dest[ax,ay]))
            
            j+=stride
            ay+=stride
        i+=stride
        ax+=stride
        j=r
        ay=say # ,,,.

    return dest

def conv1d_same(src,kernel,ax,ay,stride):
    kw,kh=kernel.shape
    if kw<=1 and kh <=1:
        print("wrong kernel size")
        return 
    if kh==1: # n1kernel,
        r=int(kw/2)
        dst=src.copy()
        dst_full=np.pad(dst,((r,r),(0,0)),constant_values=(0,0),mode='constant') # 2r0
        dst_full=dst_full.astype(np.float32)
        #dest = dst_full.copy()
        dest=np.zeros(dst_full.shape,np.float32)
        iw,ih=dest.shape
        #odds=(kw%2==1)?1:0
        if kw%2==1:
            odds=1
        else:
            odds=0
        i=j=r
        sax=ax
        say=ay
        debug_cnt=0
        while i<iw-2*r-odds:#iw+2*r-r-odds-1:
            j=r # iterator
            while j<ih-2*r:#ih-r-1:
                #if debug_cnt<10:
                #    print("[i:%d,j:%d]origin dest[%d,%d]:%d" %(i,j,ax,ay,dest[ax,ay]))
                #    print("kernel:",kernel)
                #    print("dst_full[%d:%d,%d:%d]:%s" %((i-r),(i+r+odds),(j-r),(j+r+odds),dst_full[(i-r):(i+r+odds),(j-r):(j+r+odds)]))
                #temp=kernel*dst_full[(i-r):(i+r+odds),j]
                #dest[ax,ay]=np.sum(temp)                
                temp=0
                for ki in range(2*r+odds):
                    temp+=kernel[ki]*dst_full[i][j+ki]
                    #if debug_cnt<10:
                    #    print("kernel[%d]:%s, dst_full[%d]:%s"%(ki,kernel[ki],(j+ki),dst_full[i][j+ki]))
                #print("ax:%d,ay:%d,temp:%d"%(ax,ay,temp))
                dest[ax,ay]=temp
                #if debug_cnt<10:
                #    debug_cnt+=1
                #    print("[i:%d,j:%d]after dest[%d,%d]:%d" %(i,j,ax,ay,dest[ax,ay]))

                ay+=stride
                j+=stride
            i+=stride
            ay=say # 
            ax+=stride
    elif kw==1:# n
        debug_cnt=0
        r=int(kh/2)
        dst=src.copy()
        dst_full=np.pad(dst,((0,0),(r,r)),constant_values=(0,0),mode='constant') # 2r0
        dst_full=dst_full.astype(np.float32)
        #dest = dst_full.copy()
        dest=np.zeros(dst_full.shape,np.float32)
        iw,ih=dest.shape
        #odds=(kh%2==1)?1:0
        if kh%2==1:
            odds=1
        else:
            odds=0
        i=j=r
        sax=ax
        say=ay
        while j<ih-2*r-odds:#ih+2*r-r-odds-1:
            #print("j:%d,ax:%d,ay:%d" %(j,ax,ay))
            i=r #iterator
            while i<iw-2*r:#iw-r-1:
                #temp=kernel*dst_full[i,(j-r):(j+r+odds)]
                temp=0
                for ki in range(2*r+odds):
                    #if debug_cnt<10:
                    #    print("kernel[0][%d]:%s, dst_full[%d][%d]:%s"%(ki,kernel[0][ki],(i+ki),j,dst_full[i+ki][j]))
                    temp+=kernel[0][ki]*dst_full[i+ki][j]
                #if debug_cnt<10:
                #    print("temp:%s"%(temp))
                #print("ax:%d,ay:%d,temp:%d"%(ax,ay,temp))
                dest[ax,ay]=temp
                #if debug_cnt<10:
                #    debug_cnt+=1
                #    print("[i:%d,j:%d]after dest[%d,%d]:%d" %(i,j,ax,ay,dest[ax,ay]))
                ax+=stride
                i+=stride
            ax=sax
            ay+=stride
            j+=stride

    
    return dest
```

```python
def roberts_edge(src):
    # roberts
    R1=np.array([[1,0],[0,-1]],np.float32)
    R2=np.array([[0,1],[-1,0]],np.float32)
    IconR1=conv2d_same(src,R1,0,0,stride=1)
    IconR2=conv2d_same(src,R2,0,1,stride=1)
    ##print(IconR1)
    IconR1=np.abs(IconR1)
    IconR1=IconR1.astype(np.uint8)
    IconR2=np.abs(IconR2)
    IconR2=IconR2.astype(np.uint8)
    show_img(IconR1,"edge45",1,3,1)
    show_img(IconR2,"edge135",1,3,2)
    #cv2.imshow("edge45",IconR1)
    #cv2.imshow("edge135",IconR2)
    edge_all=np.sqrt(np.power(IconR1,2.0)+np.power(IconR2,2.0))
    edge_all=np.round(edge_all)
    edge_all[edge_all>255]=255
    edge_all=edge_all.astype(np.uint8)
    show_img(edge_all,"edgeall",1,3,3)
    #cv2.imshow("edgeall",edge_all)

roberts_edge(grayImage)
#print(grayImage)
```

```python
from scipy import signal
def roberts_with_conv2d_api(src):
    R1=np.array([[1,0],[0,-1]],np.float32)
    R2=np.array([[0,1],[-1,0]],np.float32)
    IconR1=signal.convolve2d(src,R1,mode='full',boundary='fill',fillvalue=0)
    IconR2=signal.convolve2d(src,R2,mode='full',boundary='fill',fillvalue=0)
    ##print(IconR1)
    IconR1=np.abs(IconR1)
    IconR1=IconR1.astype(np.uint8)
    IconR2=np.abs(IconR2)
    IconR2=IconR2.astype(np.uint8)
    show_img(IconR1,"edge45",1,3,1)
    show_img(IconR2,"edge135",1,3,2)
    #cv2.imshow("edge45",IconR1)
    #cv2.imshow("edge135",IconR2)
    edge_all=np.sqrt(np.power(IconR1,2.0)+np.power(IconR2,2.0))
    edge_all=np.round(edge_all)
    edge_all[edge_all>255]=255
    edge_all=edge_all.astype(np.uint8)
    show_img(edge_all,"edgeall",1,3,3)
    #cv2.imshow("edgeall",edge_all)
    

roberts_with_conv2d_api(grayImage)
```

```python
def prewitt(src):
    # 1. ,
    ones_y=np.array([[1],[1],[1]],np.float32)
    #ones_y=ones_y/np.sum(ones_y)
    #print(src)
    src_smth_y=conv1d_same(src,ones_y,1,0,stride=1)
    diff_x=np.array([[1,0,-1]],np.float32)
    #diff_x=diff_x/np.sum(diff_x)
    src_diff_x=conv1d_same(src_smth_y,diff_x,0,1,stride=1)
    src_diff_x=np.abs(src_diff_x)
    src_diff_x[src_diff_x>255]=255
    src_diff_x=src_diff_x.astype(np.uint8)
    #print(src_diff_x)
    show_img(src_diff_x,"prewitt X",1,3,1)

    # 2. ,
    ones_x=np.array([[1,1,1]],np.float32)
    src_smth_x=conv1d_same(src,ones_x,0,1,stride=1)
    diff_y=np.array([[1],[0],[-1]],np.float32)
    src_diff_y=conv1d_same(src_smth_x,diff_y,1,0,stride=1)
    src_diff_y=np.abs(src_diff_y)
    src_diff_y[src_diff_y>255]=255
    src_diff_y=src_diff_y.astype(np.uint8)
    show_img(src_diff_y,"prewitt Y",1,3,2)
    
    # 3. sqrt
    
    all_img=0.5*src_diff_x+0.5*src_diff_y
    all_img[all_img>255]=255
    show_img(all_img,"all img",1,3,3)
    
prewitt(grayImage)
```

```python
def prewitt_conv_api(src):
    ones_y=np.array([[1],[1],[1]],np.float32)
    src_smth_y=signal.convolve2d(src,ones_y,mode='same',boundary='symm')
    diff_x=np.array([[1,0,-1]],np.float32)
    src_diff_x=signal.convolve2d(src_smth_y,diff_x,mode='same',boundary='symm')
    show_img(src_diff_x,"prewitt",1,2,1)

prewitt_conv_api(grayImage)
```

```python
def pascalSmooth(n): 
    pascalSmooth=np.zeros([1,n],np.float32)
    for i in range(n):
        pascalSmooth[0][i]=math.factorial(n-1)/(math.factorial(i)*math.factorial(n-1-i)) # 
    return pascalSmooth
pascalSmooth(4)
```

```python
def pascalDiff(n):
    if n<=0:
        print("n should bigger than 0")
        return
    pascalDiff=np.zeros([1,n],np.float32)
    pascal_prev=pascalSmooth(n-1)
    for i in range(n):
        if i==0:
            pascalDiff[0][i]=pascal_prev[0][i]
        elif i==n-1:
            pascalDiff[0][i]=-pascal_prev[0][i-1] # n-10,.
        else:
            pascalDiff[0][i]=pascal_prev[0][i]-pascal_prev[0][i-1] # n-1,
    return pascalDiff

A=pascalDiff(4)
#print(A.shape)
#print(A)
```

```python
#sobel kerne.sobel.kernel.
def getSobelKernel(n):
    pascalSmoothKernel=pascalSmooth(n)
    pascalDiffKernel=pascalDiff(n)
    cx=cy=int(n/2)
    sobelKernel_x=conv2d_same(pascalSmoothKernel.transpose(),pascalDiffKernel,cx,cy,stride=1)
    sobelKernel_y=conv2d_same(pascalSmoothKernel,pascalDiffKernel,cx,cy,stride=1)
    return (sobelKernel_x,sobelKernel_y)

getSobelKernel(4)
```

```python
def sobel(src,n):
    pascal_smth=pascalSmooth(n)
    pascal_diff=pascalDiff(n)
    print(pascal_smth.transpose().shape)
    print(pascal_smth.shape)
    # ,
    src_smth_y=conv1d_same(src,pascal_smth.transpose(),int(n/2),0,stride=1)
    src_diff_x=conv1d_same(src_smth_y,pascal_diff,0,int(n/2),stride=1)
    # ,
    src_smth_x=conv1d_same(src,pascal_smth,0,int(n/2),stride=1)
    src_diff_y=conv1d_same(src_smth_x,pascal_diff.transpose(),int(n/2),0,stride=1)
    src_all=np.sqrt(np.power(src_diff_x,2.0)+np.power(src_diff_y,2.0))
    src_all[src_all>255]=255
    src_all=src_all.astype(np.uint8)
    #show_img(src_all,"sobel",1,1,1)
    return src_all

src_all=sobel(grayImage,4)
show_img(src_all,"sobel with 4",1,1,1)
```

```python
def get_sobel_xy(src,n):
    if n<=0:
        print("n should bigger than 0")
        return 
    pascal_smth=pascalSmooth(n)
    pascal_diff=pascalDiff(n)
    print(pascal_smth.transpose().shape)
    print(pascal_smth.shape)
    # ,
    src_smth_y=conv1d_same(src,pascal_smth.transpose(),int(n/2),0,stride=1)
    src_diff_x=conv1d_same(src_smth_y,pascal_diff,0,int(n/2),stride=1)
    # ,
    src_smth_x=conv1d_same(src,pascal_smth,0,int(n/2),stride=1)
    src_diff_y=conv1d_same(src_smth_x,pascal_diff.transpose(),int(n/2),0,stride=1)
    
    return (src_diff_x,src_diff_y)
```

```python
def scharr(src):
    # 
    scharr_x=np.array([[3,0,-3],[10,0,-10],[3,0,-3]],np.float32)
    scharr_x=scharr_x/np.sum(np.abs(scharr_x))
    scharr_y=np.array([[3,10,3],[0,0,0],[-3,-10,-3]],np.float32)
    scharr_y=scharr_y/np.sum(np.abs(scharr_y))
    src_x=conv2d_same(src,scharr_x,1,1,stride=1)
    src_y=conv2d_same(src,scharr_y,1,1,stride=1)
    
    # 
    src_x=np.abs(src_x)
    src_x[src_x>255]=255
    src_x=src_x.astype(np.uint8)
    show_img(src_x,"img in x",2,4,1)
    
    src_y=np.abs(src_y)
    src_y[src_y>255]=255
    src_y=src_y.astype(np.uint8)
    show_img(src_y,"img in y",2,4,2)
    
    src_all=0.5*src_x+0.5*src_y
    src_all[src_all>255]=255
    show_img(src_all,"img all",2,4,3)
    
    src_sqrt=np.sqrt(np.power(src_x,2.0)+np.power(src_y,2.0))
    src_sqrt[src_sqrt>255]=255
    show_img(src_sqrt,"img sqrt",2,4,4)
    
    # 45135
    scharr_45=np.array([[0,3,10],[-3,0,3],[-10,-3,0]],np.float32)
    scharr_45=scharr_45/np.sum(np.abs(scharr_45))
    scharr_135=np.array([[10,3,0],[3,0,-3],[0,-3,-10]],np.float32)
    scharr_135=scharr_135/np.sum(np.abs(scharr_135))
    
    src_45=conv2d_same(src,scharr_45,1,1,stride=1)
    src_135=conv2d_same(src,scharr_135,1,1,stride=1)
    
    # 
    src_45=np.abs(src_45)
    src_45[src_45>255]=255
    src_45=src_45.astype(np.uint8)
    show_img(src_45,"img in 45",2,4,5)
    
    src_135=np.abs(src_135)
    src_135[src_135>255]=255
    src_135=src_135.astype(np.uint8)
    show_img(src_135,"img in 135",2,4,6)
    
    src_all_d=0.5*src_45+0.5*src_135
    src_all_d[src_all_d>255]=255
    show_img(src_all_d,"img all d",2,4,7)
    
    src_sqrt_d=np.sqrt(np.power(src_45,2.0)+np.power(src_135,2.0))
    src_sqrt_d[src_sqrt_d>255]=255
    show_img(src_all_d,"img sqrt d",2,4,8)

scharr(grayImage)
```

```python
# non max suppression
# dx: 
# dy: 

def nms_default(dx,dy):
    # 
    edge=np.sqrt(np.power(dx,2.0),np.power(dy,2.0))
    edge[edge>255]=255
    gradiant_angle=np.zeros(edge.shape,np.float32)
    nms_edge=np.zeros(edge.shape,np.float32)
    w,h=edge.shape
    for r in range(1,h-1): # 
        for c in range(1,w-1):
            angle=math.atan2(dy[r][c],dx[r][c])/math.pi*180 #
            gradiant_angle[r][c]=angle # r,c
            if (np.abs(angle)<22.5 or np.abs(angle)>157.5): # 
                if edge[r][c]>edge[r][c-1] and edge[r][c]>edge[r][c+1]:
                    nms_edge[r][c]=edge[r][c] #.
            if (angle>=22.5 and angle<67.5 or ((-angle)>=122.5 and (-angle)<157.5)): # 
                if edge[r][c]>edge[r-1][c-1] and edge[r][c]>edge[r+1][c+1]:
                    nms_edge[r][c]=edge[r][c] 
            if (np.abs(angle)>=67.5 and np.abs(angle)<122.5): # 
                if edge[r][c]>edge[r-1][c] and edge[r][c]>edge[r][c+1]:
                    nms_edge[r][c]=edge[r][c] 
            if (angle>=122.5 and angle<157.5 or ((-angle)>=22.5 and (-angle)<67.5)): # 
                if edge[r][c]>edge[r-1][c+1] and edge[r][c]>edge[r+1][c-1]:
                    nms_edge[r][c]=edge[r][c] 
    return nms_edge
```

```python
# 
# ,.
def trace_edge(nms_edge,edge,lth,r,c,rows,cols):
    if edge[r][c]==0: #
        edge[r][c]=255
        # low threash,,
        for i in range(-1,2):
            for j in range(-1,2):
                if valid_param(r+i,c+j,rows,cols) and (nms_edge[r+i][c+j]>=lth):
                    trace_edge(nms_edge,edge,lth,r+i,c+j,rows,cols)
        
```

```python
def valid_param(r,c,rows,cols):
    if r>=0 and r<rows and c>=0 and c<cols:
        return True
    else:
        return False
    
```

```python
def hysteresisThreshold(nms_edge,lth,hth):
    w,h=nms_edge.shape
    edge=np.zeros(nms_edge.shape,np.float32)
    for r in range(1,h-1):
        for c in range(1,w-1):
            if nms_edge[r][c]>=hth:
                trace_edge(nms_edge,edge,lth,r,c,w,h)
            if nms_edge[r][c]<lth:
                edge[r][c]=0
    
    return edge
```

### LoG 

1. :

$$gauss(x,y,\sigma )=\frac{1}{2\pi\sigma^2}exp(-\frac{x^2+y^2}{2\sigma^2})$$

2. :

    1. 
    ![](https://wikimedia.org/api/rest_v1/media/math/render/svg/f81035fd6bb5b1a76a2732b58b7056a24898c14d)
    2. :
    $$\delta^2(gauss(x,y,\sigma))=\frac{\delta^2(guass(x,y,\sigma))}{\partial^2x}+\frac{\delta^2(guass(x,y,\sigma))}{\partial^2y}=\frac{1}{2\pi\sigma}(\frac{x^2+y^2}{\sigma^2}-2)exp(-\frac{x^2+y^2}{2\sigma^2})$$
    
3. LoG
$$h:[1->H-1]$$
$$w:[1->W-1]$$
$$LoG_{H*W}=[\delta ^2guass(h-\frac{H-1}{2},w-\frac{W-1}{2},\sigma)]$$

```python
def createLoGKernel(sigma,kw,kh):
    if kw!=kh:
        print("kernel width should equals its height")
        return 
    if kw%2!=1:
        print("kernel width and height should be odds")
        return 
    sigma2=np.power(sigma,2.0)
    kernel=np.zeros((kw,kh),np.float32)
    #print(kernel)
    for w in range(kw):
        for h in range(kh):
            i=w-(kw/2-1)
            j=h-(kh/2-1)
            #print("i:%d,j:%d"%(i,j))
            kernel[w,h]=((np.power(i,2.0)+np.power(j,2.0))/sigma2 -2)*np.exp(-1*(np.power(i,2.0)+np.power(j,2.0))/(2.0*sigma2))
    
    #print(kernel)
    return kernel

k=createLoGKernel(2.0,3,3)
k=k/np.sum(k)
print(k)
```

```python
def LoG(src,sigma,kw,kh):
    LoGkernel=createLoGKernel(sigma,kw,kh)
    LoGkernel_norm=LoGkernel/np.sum(LoGkernel)
    ax=int((kw-1)/2)
    ay=int((kh-1)/2)
    dest=conv2d_same(src,LoGkernel_norm,ax,ay,stride=1)
    return dest

sigma=2.0
ksize=13
dst_image=LoG(grayImage,2.0,13,13)
#dst_image=np.abs(dst_image)          # abs,.
#dst_image=dst_image.astype(np.uint8)
dst_image[dst_image>255]=255
dst_image[dst_image<0]=0
show_img(dst_image,"LoG image sigma%s kernel%s"%(sigma,ksize),1,1,1)
    
```
